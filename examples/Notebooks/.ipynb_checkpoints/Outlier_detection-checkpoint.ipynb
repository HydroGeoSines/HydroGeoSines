```python
# extract data object
csiro_data = csiro_site.data

# these are our groundwater record locations
print("List of GW locations: {}".format(csiro_data.hgs.filters.get_gw_locs))

# lets only use one of them for now
gw_locs = csiro_data.hgs.filters.get_gw_locs
locA_data = csiro_data.loc[csiro_data.location==gw_locs[0]]
```

## Outliers
There are three major types of outliers that can be detected in this example of groundwater records ('Loc_A') in the CSIRO dataset:

- **point wise:** These can be extreme values or spikes in our data set, for example due to reading errors of the sensor. Globally, they can be be detected using statistical values such as mean, max, min, quantilies and standard deviation. In case of the groundwater record from location 'A_all' two such spikes are found at the start of the data set, in January 2001. We can replace these values with the median of the remaining data.

- **pattern wise (collective values):** Here we see a shift of a whole section of the reading between February and March of the record. In this case the pattern is encountered together with a change in sampling rate beforehand. These might occure due to maintainance of the sensor.

- **system wise:** sharp rises or falls -> instationarity of the system, e.g. due to pumping or groundwater recharge.


### Detect outliers
For anomalie detection (outliers) in our time series we can use a rolling window to calculate statistics for a given period. A very convenient meaure is the percent change of a given value, compared to the mean of all consecutive values within a given window. Compared to the raw data, the transform data is now stationary, which can be very convenient in time series analysis. Stationarity is given when our data has a constant mean and standard deviation over time.

```python
# check stationarity of our raw data
outliers.stationary(locA_data)
```

### Find point-wise outliers
Now, lets use a rolling window approach to calculate the percent change as described earlier. In this case we define outliers using the standard deviation as our statistical metric. Assuming a normal distribution, 99.7% of data observed lies within 3 standard deviations of the mean. Any value exceeding this threshold is labeled an outlier. 

```python
# in our example, we use a rolling window lenght of 24H
df = outliers.plot_outliers(locA_data, window = "24H",cat="GW")
```
